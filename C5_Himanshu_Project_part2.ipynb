```python
# Importing the required libraries
import matplotlib.pyplot as plt          
from tabulate import tabulate               
import matplotlib as mat                     
import seaborn as sns                    
import pandas as pd                         
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load data
df = pd.read_csv('DS1_C5_S5_Hotel Dataset.csv')

# Null Value Treatment
for column in df.columns:
    if df[column].isnull().sum() > 0:  # only treat if nulls are present
        if df[column].dtype == 'object':  # categorical
            mode_val = df[column].mode()[0]
            df[column].fillna(mode_val, inplace=True)
        else:  # numerical
            median_val = df[column].median()
            df[column].fillna(median_val, inplace=True)

# Drop Duplicates
df.drop_duplicates(inplace=True)
```

### Task 1: Customer Satisfaction

```python
# 1.1 Distribution of number of special requests
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='total_of_special_requests', palette='Set2')
plt.title('Distribution of Special Requests')
plt.xlabel('Number of Special Requests')
plt.ylabel('Count')
```

```python
# 1.2 Special requests vs. ADR
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='total_of_special_requests', y='adr', palette='Set3')
plt.title('ADR vs. Special Requests')
plt.xlabel('Number of Special Requests')
plt.ylabel('Average Daily Rate (ADR)')
```

```python
# 1.3 Room preferences by customer type
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='reserved_room_type', hue='customer_type', palette='husl')
plt.title('Room Type Preferences by Customer Type')
plt.xlabel('Reserved Room Type')
plt.ylabel('Count')
plt.legend(title='Customer Type')
```

### Task 2: Marketing and Sales Optimization

```python
# 2.1 Market segment contribution to bookings
plt.figure(figsize=(8,6))
sns.countplot(data=df, y='market_segment', order=df['market_segment'].value_counts().index, palette='coolwarm')
plt.title('Bookings by Market Segment')
plt.xlabel('Count')
plt.ylabel('Market Segment')
```

```python
# 2.2 ADR by meal type (proxy for amenities/services)
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='meal', y='adr', palette='pastel')
plt.title('ADR by Meal Type')
plt.xlabel('Meal Plan')
plt.ylabel('Average Daily Rate (ADR)')
```

```python
# 2.3 Distribution channel analysis
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='distribution_channel', palette='Set1')
plt.title('Distribution of Bookings by Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Count')
```
